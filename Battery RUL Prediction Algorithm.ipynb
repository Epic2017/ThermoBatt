```python
#Predict remaining number of cycles for battery discharge.

import pandas as pd

my_data = pd.read_csv('dis_ele_with_csv.csv')

my_data.describe()

#Defining maximum sustainable usage at 70% of original capacity. This dataset ends tests when batteries have approached end of life.

for i in my_data.columns:
    print(i)
```

```python
#Isolate features for testing

RUL_test = my_data[['Voltage_measured','Current_measured','Temperature_measured','ambient_temperature','Corrected_time','Cycle_remaining']]
RUL_test.head()
```

Prediction with standard features

```python
#Start out with a simple LinearRegression model (base). Import various regression packages: 

from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

#Association between temp data and RUL predictions.

Y = RUL_test["Cycle_remaining"]
X = RUL_test.drop(['Cycle_remaining'], axis=1) #

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25)

regressor = LinearRegression()
regressor.fit(X_train, Y_train)

Y_pred_train = regressor.predict(X_train)

Y_pred_test = regressor.predict(X_test)

#Develop upon error statistics for further analysis.


train_review = pd.DataFrame({"Model Prediction": Y_pred_train,
             "Expected/Real values": Y_train})

abs_error = (abs(Y_pred_test - Y_test))
print(abs_error)

test_review = pd.DataFrame({"Model Prediction": Y_pred_test,
             "Expected/Real values": Y_test, "Absolute Error": abs_error})

my_error = mean_squared_error(Y_test, Y_pred_test, squared=False)

r2 = r2_score(Y_test, Y_pred_test)
print('Error stats:' , my_error, r2)
```
