```python
#import libraries and view dataset
import numpy as np
import pandas as pd
from pandas import Series
from numpy.random import randn
import csv #csv library col-name
import matplotlib.pyplot as plt
import seaborn as sns

#import data from google drive
from google.colab import drive
drive.mount('/content/drive')

# read charge, discharge, and impedance data for B0005 and create pandas dataframes for each
charge_data = pd.read_csv('/content/drive/MyDrive/BatteryProjectData/MyCleanedData/charge_data_B0005.csv')
discharge_data = pd.read_csv('/content/drive/MyDrive/BatteryProjectData/MyCleanedData/discharge_data_B0005.csv')
impedance_data = pd.read_csv('/content/drive/MyDrive/BatteryProjectData/MyCleanedData/impedance_data_B0005.csv')
```

```python
# Assuming discharge_data is your DataFrame
# Group by 'cycle_number' and then take the first occurrence of each group
first_instances = discharge_data.groupby('cycle_number').head(1)

attrib=['cycle_number','Time','Capacity']
dis_ele= first_instances[attrib]
print(dis_ele.head())
```

```python
#plotting the capacity vs cycle graph from discharge_data
%matplotlib inline

plot_df = dis_ele.loc[(dis_ele['cycle_number']>=1),['cycle_number','Capacity']]
sns.set_style("darkgrid")
plt.figure(figsize=(12, 8))
plt.plot(plot_df['cycle_number'], plot_df['Capacity'])
#Draw threshold
plt.plot([0.,len(dis_ele)], [1.4, 1.4])
plt.ylabel('Capacity')
# make x-axis ticks legible
adf = plt.gca().get_xaxis().get_major_formatter()
plt.xlabel('cycle')
plt.title('Discharge B0005')
```

```python
'''Calculate The state of health of battery No.05:

SOH(τ)=C(τ)/C(0)
where C(0) is the capacity value at the initial stage of Li-ion battery, and C(τ) is the capacity value at time τ (it is usually the index of cycle number).'''

#create & calculate a new feature called state of health and add it to the dataframe
attrib=['cycle_number','Time','Capacity']
dis_ele=dis_ele[attrib]
C=dis_ele['Capacity'][0]
#print(dis_ele['Capacity'][0])
HoS=0
for i in range(len(dis_ele)):
    dis_ele['SoH']=(dis_ele['Capacity'])/C
print(dis_ele.head(5))
print(dis_ele.tail(5))
```

```python
import pandas as pd

# Load your dataset into a Pandas DataFrame
#import data from google drive
from google.colab import drive
drive.mount('/content/drive')
df = pd.read_csv('/content/drive/MyDrive/BatteryProjectData/MyCleanedData/all_data_fixed.csv')  # Replace with the actual file path

# Initialize a variable to track the previous cycle type
previous_type = None

# Initialize a list to hold the indices of the first instances of discharge cycles
first_discharge_indices = []

# Iterate through the DataFrame
for index, row in df.iterrows():
    # Check if the current cycle is a discharge cycle and the previous cycle was not a discharge cycle
    if row['cycle_type'] == 'discharge' and previous_type != 'discharge':
        first_discharge_indices.append(index)

    # Update the previous cycle type
    previous_type = row['cycle_type']

# Filter the DataFrame to keep only the first instances of discharge cycles
filtered_df = df.loc[first_discharge_indices]

# Check the filtered DataFrame
print(filtered_df.head())

# Check if filtered_df is empty
if len(filtered_df) == 0:
    print("No discharge cycles found.")
else:
    # Get the first element of the 'Capacity' column
    C = filtered_df['Capacity'].iloc[0]

    # Calculation for State of Health (SoH)
    filtered_df['SoH'] = filtered_df['Capacity'] / C

    # Save the filtered DataFrame to a new CSV file
    filtered_df.to_csv('/content/drive/MyDrive/BatteryProjectData/MyCleanedData/dis_ele.csv', index=False)  # Replace with the desired output file path
```
